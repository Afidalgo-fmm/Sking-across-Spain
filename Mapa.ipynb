```python
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

```python
data = pd.read_csv('data/Datos.csv', index_col = 0)
data1 = pd.read_csv('data/Restaurantes.csv', index_col = 0)
rest = pd.DataFrame(data1)
df = pd.DataFrame(data)
```

```python
#type(df.iloc[0].lat)
```

```python
inicial_lat = 40.463667
inicial_long = -3.74922
esp_map = folium.Map(location=[inicial_lat,inicial_long], zoom_start=6, tiles = 'Stamen Terrain')
esp_map;
```

```python
'''icono = Icon(color="green",
             opacity=0.3,
             prefix = "fa",
             icon="paw",
             icon_color="black")'''
```

```python
# Localizacion e Iconos de estaciones
for i,row in df.iterrows():
    #print([float(row.lat),float(row.long)])
    #print(row.estacion)
    marker1 = folium.Marker(location=[float(row.lat),float(row.long)], tooltip= row.estacion, icon=folium.Icon(icon='glyphicon glyphicon-asterisk'))
    marker1.add_to(esp_map)
    
```

```python
rest.info()
```

```python
# Localizacion e Iconos para restaurantes
for i,row in rest.iterrows():
    #print([float(row.lat),float(row.long)])
    #print(row.estacion)
    marker1 = folium.Marker(location=[float(row.lat),float(row.lng)], tooltip= row.nombre_rest, icon=folium.Icon(icon='glyphicon glyphicon-cutlery'))
    marker1.add_to(esp_map)
```
