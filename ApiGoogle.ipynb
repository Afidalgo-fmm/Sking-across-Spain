```python
def restaurantfinder(lat, long):
    re = cliente.places_nearby(location = f'{lat},{long}', radius= 5000, open_now = False, type = 'restaurant')
    return re
```

```python
# Sonia's ideas ideitas
megalista_restaurantes =[]
for i in range(len(respuesta)):
    estacion = restauranteasociado[i]
    if len(respuesta[i]['results']) != 0:
        for j in range(len(respuesta[i]['results'])): # j es el posicion dentro del results de mi respuesta
            new_dict = {'nombre_est':estacion, 'coord':None, 'nombre_rest':None,'adress':None,'rating':None}
            #para las coordenadas
            try:
                new_dict["coord"] = [respuesta[i]['results'][j]['geometry']['location']['lat'],respuesta[i]['results'][j]['geometry']['location']['lng']]

            except:
                try:
                    new_dict["coord"] =[respuesta[i]['results'][j]['viewport']['northeast']['lat'],respuesta[i]['results'][j]['viewport']['northeast']['lng']]
                except:
                    pass
            #para el nombre
            try:
                new_dict['nombre_rest'] = [respuesta[i]['results'][j]['name']]
            except:
                pass
            #para adress
            try:
                new_dict['adress'] = respuesta[i]['results'][j]['vicinity']
            except:
                pass
            #rating
            try:
                new_dict['rating'] = respuesta[i]['results'][j]['rating']
            except:
                pass
            #por ultimo appendeamos el diccionario a un megalista
            megalista_restaurantes.append(new_dict)
    else:
        new_dict = {'nombre_est':estacion, 'coord':None, 'nombre_rest':None,'adress':None,'rating':None}
        megalista_restaurantes.append(new_dict)
```
