```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config Inlinebackend.figure_format= 'retina'
sns.set_context("poster")
sns.set(rc={"figure.figsize": (12.,6.)})
sns.set_style("whitegrid")
```

```python
data = pd.read_csv('data/Datos.csv', index_col = 0)
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
ncount = len(df.Snowpark)
ax = sns.countplot(x=df.Snowpark, hue= 'Snowpark', data = df)
plt.title('Estaciones con o sin Snnowpark')
```

```python
ax = sns.countplot(x=df.Familiar, hue= 'Familiar', data = df)
plt.title('ESTACIONES PARA FAMILIAS')
```

```python
datos= df.Familiar.value_counts()
datos.plot.pie(autopct="%.1f%%")
```

```python
datos= df.Snowpark.value_counts()
datos.plot.pie(autopct="%.1f%%")
```
